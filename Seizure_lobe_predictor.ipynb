```python
import numpy as np
import matplotlib.pyplot as plt
import mne
from tvb.simulator.lab import *
from pyprep.prep_pipeline import PrepPipeline

raw = mne.io.read_raw_fif('eeg_sim_raw.fif')
# raw.plot(scalings=1e-1)
sample_rate = raw.info["sfreq"]

mon_EEG = monitors.EEG.from_file()
ch_names = mon_EEG.sensors.labels
ch_locations = mon_EEG.sensors.locations
montage = mne.channels.make_dig_montage(dict(zip(ch_names, ch_locations/1000)))

# montage_std = mne.channels.make_standard_montage("standard_1020")
# montage_std.plot(show_names=True)

raw_copy = raw.copy()
prep_params = {'ref_chs': ch_names,
               'reref_chs': ch_names,
               'line_freqs': np.arange(50, sample_rate/2, 50)}
prep = PrepPipeline(raw_copy, prep_params, montage)
prep.fit()
```
